```python
import numpy as np
import gymnasium as gym
import matplotlib.pyplot as plt
```

#### calculate discounted rewards ex

```python
exp_rewards_strategy_1 = np.array([3, 2, -1, 5])

DISCOUNT_FACTOR = 0.9

# Compute discount
discounts_strategy_1 = np.array([DISCOUNT_FACTOR ** i for i in range(len(exp_rewards_strategy_1))])

# Compute the discounted return
discounted_return_strategy_1 = np.sum(discounts_strategy_1 * exp_rewards_strategy_1)

print(f"The discounted return of the first strategy is {discounted_return_strategy_1}")
```

```python
exp_rewards_strategy_2 = np.array([6, -5, -3, -2])

DISCOUNT_FACTOR_2 = 0.9

# Compute discount
discounts_strategy_2 = np.array([DISCOUNT_FACTOR_2 ** i for i in range(len(exp_rewards_strategy_2))])

# Compute the discounted return
discounted_return_strategy_2 = np.sum(discounts_strategy_2 * exp_rewards_strategy_2)

print(f"The discounted return of the first strategy is {discounted_return_strategy_2}")
```

### mountain car gym

```python
# Create the env
env = gym.make('MountainCar', render_mode='rgb_array')

# Get the inital state
initial_state, info = env.reset(seed=42)

position = initial_state[0]
velocity = initial_state[1]

print(f"The position of the car along the x-axis is {position} (m)")
print(f"The velocity of the car is {velocity} (m/s)")
```

```python
def render():
    state_image = env.render()
    plt.imshow(state_image)
    plt.show()
```

```python
render()
```

### Frozen Lake

```python
env = gym.make('FrozenLake', render_mode='rgb_array')

initial_state, info = env.reset(seed=42)
```

```python
# Define the sequence of actions
actions = [1, 1, 2, 2, 1, 2]

for action in actions:
    # execute each action
    state, reward, terminated, _, _ = env.step(action)
    # render the env
    render()
    if terminated:
        print("actions have been terminated")
```
