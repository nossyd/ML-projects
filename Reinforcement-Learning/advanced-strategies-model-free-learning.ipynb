## Expected SARSA

```python
def update_q_table_expected_sarsa(state, action, next_state, reward, Q, alpha, gamma):
    old_value = Q[state, action]
    expected_q = np.mean(Q[state])

    Q[state, action] = (1 - alpha) * old_value + alpha * (reward + gamma * expected_q)

    return Q
```
