```python
def epsilon_greedy(state, Q, epsilon):
    # implement condition to explore
    if np.random.rand() < epsilon:
        # choose random action
        action = env.action_space.sample()
    else:
        # choose best action according to Q table
        action = np.argmax(Q[state, :])
    
    return action
```

## Multi-arm Bandits

```python
# create a simulated multi-arm bandit
def create_multi_arm_bandit(n_bandits, n_iterations):
    # generate true bandit probs
    true_bandit_probs = np.random.rand(n_bandits)
    # init arrays to hold count and values for each bandit
    counts = np.zeros(n_bandits)
    values = np.zeros(n_bandits)
    # init arrays to hold rewards and selected arms for each episode
    rewards = np.zeros(n_iterations)
    selected_arms = np.zeros(n_iterations)

    return true_bandit_probs, counts, values, rewards, selected_arms
```

```python
# init params
epsilon = 1.0
min_epsilon = 0.01
epsilon_decay = 0.999
n_bandits = 10
n_iterations = 1000
```
