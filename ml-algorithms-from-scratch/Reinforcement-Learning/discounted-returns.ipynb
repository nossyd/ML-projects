```python
import numpy as np
def discount_returns(rewards, gamma):
    """
    Compute the discounted return at every timestep.
    """
    # Write code here
    r = np.array(rewards, dtype=np.float64)
    returns = np.empty_like(r)
    running = 0.0

    for i in range(len(r) -1, -1, -1):
        running = r[i] + gamma * running
        returns[i] = running

    return returns
```

#### test case #1

Input:
rewards = [1, 1, 1], gamma = 1.0

Output:
[3.0, 2.0, 1.0]
With gamma = 1 there is no discounting. G[0] = 1 + 1 + 1 = 3, G[1] = 1 + 1 = 2, G[2] = 1.

```python
rewards1 = [1, 1, 1]
gamma1 = 1.0

returns1 = discount_returns(rewards=rewards1, gamma=gamma1)
returns1
```

#### test case #2

Input:
rewards = [0, 0, 0, 10], gamma = 0.9

Output:
[7.29, 8.1, 9.0, 10.0]
Only the last timestep has a reward. The return at earlier steps is that reward discounted by gamma raised to the number of steps away: G[0] = 10 * 0.9^3 = 7.29.

```python
rewards2 = [0, 0, 0, 10]
gamma2 = 0.9

returns2 = discount_returns(rewards=rewards2, gamma=gamma2)
returns2
```
