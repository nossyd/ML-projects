```python
import numpy as np

def compute_advantage(rewards, V, gamma):
    """
    Returns: A (NumPy array of advantages)
    """
    r = np.array(rewards, dtype=np.float64)
    V = np.array(V, dtype=np.float64)
    G_t = np.empty_like(r)
    A_t = np.empty_like(r)
    running = 0.0

    for i in range(len(r) - 1 , -1, -1):
        running = r[i] + gamma * running
        G_t[i] = running

        A_t[i] = G_t[i] - V[i]
        
    return A_t
```

#### test case #1

Input: states=[0,1,2], rewards=[1,2,3], V=[0.5,1,1.5], γ=1

Output: A = [5.5, 4.0, 1.5]

```python
states1=[0,1,2]
rewards1=[1,2,3]
V1=[0.5,1,1.5]
gamma1=1

A_t1 = compute_advantage(rewards=rewards1, V=V1, gamma=gamma1)
A_t1
```

#### test case #2
