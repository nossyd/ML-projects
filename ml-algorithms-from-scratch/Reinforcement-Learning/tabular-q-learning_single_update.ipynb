```python
import numpy as np

def q_learning_update(Q, s, a, r, s_next, alpha, gamma):
    """
    Returns: updated Q-table Q_new
    """
    Q = np.array(Q, dtype=float)
    target = r + gamma * np.max(Q[s_next])
    Q_new = Q.copy()
    Q_new[s,a] = Q_new[s,a] + alpha * (target-Q[s,a])
    return Q_new
```

#### test case #1

Input: Q=[[0,0],[0.5,1]], s=0, a=1, r=1, s_next=1, α=0.5, γ=0.9

Output: Q_new = [[0, 0.95], [0.5, 1]]

```python
Q1=[[0,0],[0.5,1]]
s1=0
a1=1
r1=1
s_next1=1
alpha1=0.5
gamma1=0.9

Q_new1 = q_learning_update(Q=Q1, s=s1, a=a1, r=r1, s_next=s_next1, alpha=alpha1, gamma=gamma1)
Q_new1
```
