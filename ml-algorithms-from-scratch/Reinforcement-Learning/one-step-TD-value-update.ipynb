```python
import numpy as np

def td_value_update(V, s, r, s_next, alpha, gamma):
    """
    Returns: updated value function V_new
    """
    V = np.array(V, dtype=float)
    V_new = V.copy()
    delta = r + gamma * V_new[s_next] - V_new[s]
    V_new[s] = V_new[s] + (alpha * delta)
    return V_new
```

#### test case #1

Input: V=[0,0,0], s=0, r=1, s_next=1, α=0.5, γ=0.9

Output: V_new = [0.5, 0.0, 0.0]

```python
V1=[0,0,0]
s1=0
r1=1
s_next1=1
alpha1=0.5
gamma1=0.9

V_new1 = td_value_update(V=V1, s=s1, r=r1, s_next=s_next1, alpha=alpha1, gamma=gamma1)
V_new1
```
