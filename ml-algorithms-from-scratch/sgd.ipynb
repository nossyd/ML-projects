```python
# loop for SGD
def sgd_loop(data, weights, lr=0.05, loops=10):
    for n in range(loops):
        weights = weights + lr * (data[n, 1] - (data[n, 0] * weights)) * data[n, 0]
    return weights
```

```python
wi = sgd_loop(random_data, random_w)
wi
```
