```python
def generator(z, W, b):
    """
    Returns: np.ndarray of shape (batch, output_dim) with tanh-activated values rounded to 4 decimals
    """
    z = np.asarray(z, dtype=np.float64)
    W = np.asarray(W, dtype=np.float64)
    b = np.asarray(b, dtype=np.float64)

    return np.round(np.tanh(z @ W + b), 4)
```

```python
def discriminator(x, W):
    """
    Returns: np.ndarray of shape (batch, 1) with probabilities rounded to 4 decimals
    """
    x = np.asarray(x, dtype=np.float64)
    W = np.asarray(W, dtype=np.float64)
    logits = x @ W
    return np.round(1 / (1 + np.exp(-(logits))),4)
```

```python
def discriminator_loss(real_probs, fake_probs):
    """Compute discriminator loss using binary cross-entropy.
    Returns: Loss value rounded to 4 decimals."""
    CLIPPED_LVL = 1e-8 
    real_probs = np.clip(np.asarray(real_probs, dtype=np.float64), CLIPPED_LVL, 1-CLIPPED_LVL)
    fake_probs = np.clip(np.asarray(fake_probs, dtype=np.float64), CLIPPED_LVL, 1-CLIPPED_LVL)

    return np.round(-np.mean(np.log(real_probs) + np.log(1-fake_probs)),4)
```

```python
def generator_loss(fake_probs):
    """Compute non-saturating generator loss.
    Returns: Loss value rounded to 4 decimals."""
    CLIPPED_LVL = 1e-8 
    fake_probs = np.clip(np.asarray(fake_probs, dtype=np.float64), CLIPPED_LVL, 1-CLIPPED_LVL)

    return np.round(-np.mean(np.log(fake_probs)),4)
```

```python
def train_gan_step(real_data, fake_data, D_W):
    """
    Returns: dict with "d_loss" and "g_loss" as float values
    """
    EPSILON = 1e-8
    
    real_data = np.asarray(real_data, dtype=np.float64)
    fake_data = np.asarray(fake_data, dtype=np.float64)
    D_W = np.asarray(D_W, dtype=np.float64)

    p_real = np.clip(1 / (1 + np.exp(-real_data@D_W)), EPSILON, 1-EPSILON)
    p_fake = np.clip(1 / (1 + np.exp(-fake_data@D_W)), EPSILON, 1-EPSILON)

    d_loss = -np.mean(np.log(p_real) + np.log(1-p_fake))
    g_loss = -np.mean(np.log(p_fake))

    return {'d_loss': np.round(d_loss, 4), 'g_loss': np.round(g_loss, 4)}
```

Output: {"d_loss": 1.2371, "g_loss": 0.6855}

```python
def detect_mode_collapse(generated_samples, threshold=0.1):
    """
    Returns: dict with "diversity_score" (float) and "is_collapsed" (bool)
    """
    generated_samples = np.asarray(generated_samples, dtype=np.float64)
    diversity_score = np.mean(np.std(generated_samples, axis=0))
    is_collapsed = diversity_score < threshold

    return {"diversity_score": float(np.round(diversity_score, 4)), "is_collapsed": bool(is_collapsed)}
```
